```python
# 创建spark_session
os.environ['PYSPARK_PYTHON']="/usr/local/anaconda3/bin/python"

def Start_Spark():
    Spark_Session = SparkSession.builder \
                    .appName("kg_spark_for_model") \
                    .enableHiveSupport() \
                    .config("spark.default.parallelism", "600") \
                    .config("spark.sql.shuffle.partitions", "3000") \
                    .config("spark.sql.broadcastTimeout","3600")\
                    .config("spark.driver.memory", "200g") \
                    .config("spark.executor.memory", "40g")\
                    .config("spark.yarn.appMasterEnv.yarn.nodemanager.container-executor.class", "DockerLinuxContainer")\
                    .config("spark.executorEnv.yarn.nodemanager.container-executor.class", "DockerLinuxContainer")\
                    .config("spark.yarn.appMasterEnv.yarn.nodemanager.docker-container-executor.image-name", 
                            "bdp-docker.jd.com:5000/wise_mart_bag:latest")\
                    .config("spark.executorEnv.yarn.nodemanager.docker-container-executor.image-name", 
                            "bdp-docker.jd.com:5000/wise_mart_bag:latest")\
                    .config("spark.sql.crossJoin.enabled", "true")\
                    .config("spark.driver.maxResultSize", "40g")\
                    .config("spark.sql.autoBroadcastJoinThreshold","104857600")\
                    .getOrCreate()
    
    return Spark_Session
```

# 预设参数

```python
from kg_lib.utils import mkdir
from kg_lib.utils import read_json_config_file

# 目标节点类型
Aim_Node_Type = 'Mobile_Node'

# 目标时间
Aim_Feature_Time = '2023-04-01'

# 一批采样的最大样本总数
sample_batch_size = 70000

#召回源数据的路径
source_data_file_path = '../../Target_Node/新高发件召回源_0508.pkl'

# 数据保存的位置
Feature_Data_From_Online_Store_dir = '../../Data/06_21-高发件召回源4月1号特征-新/'
mkdir('../../Data/')
mkdir(Feature_Data_From_Online_Store_dir)

Model_Config_dict = {}

Model_Config_dict['Data_Regenerate'] = False

Model_Config_dict['Feature_Preprocess_Type_List'] = ['Norm']

# Model_Config_dict['Meta_path_drop_list'] = ['mobile_2_company', 'company_related_to_l4_industry']
# Model_Config_dict['Meta_path_drop_list'] = ['user_pin_related_to_mobile_by_set', 'mobile_company_site', 'mobile_receive_from_site', 
#                                             'company_related_to_industry']
Model_Config_dict['Meta_path_drop_list'] = {}

# Model_Config_dict['Meta_path_Column_drop_dict'] = {'user_pin_related_to_mobile_by_express': [0], 'mobile_send_from_site': [0, 2]}
# Model_Config_dict['Meta_path_Column_drop_dict'] = {'user_pin_related_to_mobile_by_express': [0], 'mobile_send_from_site': [0, 2], 
#                                                    'company_related_to_industry': [0, 2, 4, 5, 6]}
# Model_Config_dict['Meta_path_Column_drop_dict'] = {'company_related_to_industry': [6]}
Model_Config_dict['Meta_path_Column_drop_dict'] = {}

print(Model_Config_dict)

# Feature_Data_From_Online_Config_file = './kg_config/Feature_Data_From_Online_Config/Feature_Data_From_Online_Config_2023_05_25.json'
Feature_Data_From_Online_Config_file = './kg_config/Feature_Data_From_Online_Config/Complex_Path/Feature_Data_From_Online_Config_2023_06_15_Raw_ML.json'
Feature_Data_From_Online_Config_dict = read_json_config_file(Feature_Data_From_Online_Config_file)
```

# 读取模型

```python
# Lgb_model_file = '../Model_Parameter/Lgb/05-26-14:00_Train_2023_01_01-2023_03_01_model_parameter.txt'
Lgb_model_file = '../../Model_Parameter/Lgb/06-21-14:49_Train_2023_01_01-2023_03_01_model_parameter.txt'

Mlgb = lgb.Booster(model_file = Lgb_model_file)
```

# 读取数据并预测

```python
from kg_lib.Get_Aggregated_Data_From_Online_quick import Get_Aggregated_Data_From_Online, Concat_Feature_For_ML_Model
from kg_lib.Target_Node_Dataloader import get_aim_UID_with_label_rdd, Extend_feature_time_for_aim_UID
from py4j.protocol import Py4JJavaError

spark_restart_count = 0

sample_batch_start = 0

target_node_with_proba_pd_list = []

start_time = datetime.now()
print(start_time)

while True:
    try:
        # 启动spark
        Spark_Session = Start_Spark()

        #读取召回源数据
        tmp_aim_entity_pd = pd.read_pickle(source_data_file_path)
        tmp_aim_entity_pd = tmp_aim_entity_pd[['irr_mobile_sha']].copy()
        tmp_aim_entity_pd = tmp_aim_entity_pd.rename(columns = {"irr_mobile_sha": "Mobile_Node_UID"})
        tmp_aim_entity_pd['Feature_Time'] = Aim_Feature_Time
        
        # 获得召回源样本总数
        tmp_source_node_num = tmp_aim_entity_pd.shape[0]
        print('召回源样本总数:', tmp_source_node_num)

        for tmp_range_start in range(sample_batch_start, tmp_source_node_num, sample_batch_size):
            tmp_index = tmp_range_start//sample_batch_size

            tmp_start_time = datetime.now()

            tmp_range_end = tmp_range_start + sample_batch_size
            if tmp_range_end > tmp_source_node_num:
                tmp_range_end = tmp_source_node_num

            print('当前处理样本区间', tmp_range_start, tmp_range_end, tmp_index)

            tmp_all_output_data_batch_dir = (Feature_Data_From_Online_Store_dir + str(tmp_range_start) + '-' + str(tmp_range_end) + '/')
            mkdir(tmp_all_output_data_batch_dir)
            print('目标时间数据存储文件夹', tmp_all_output_data_batch_dir)
            
            tmp_sub_aim_entity_pd = tmp_aim_entity_pd.iloc[tmp_range_start:tmp_range_end, :].reset_index(drop = True)

            # 转rdd
            tmp_node_table_schema = StructType([StructField('Mobile_Node_UID', StringType(), True),
                                                StructField('Feature_Time', StringType(), True)])
            tmp_aim_entity_rdd = Spark_Session.createDataFrame(tmp_sub_aim_entity_pd, tmp_node_table_schema)

            tmp_aim_UID_info_dict = {'Node_Type': 'Mobile_Node',
                                        'Data': tmp_aim_entity_rdd,
                                        'Data_pd': tmp_sub_aim_entity_pd}

            tmp_output_data_label_file = tmp_all_output_data_batch_dir + 'Target_Node.pkl'

            if not os.path.exists(tmp_output_data_label_file):
                tmp_sub_aim_entity_pd.to_pickle(tmp_output_data_label_file)

            print('完成目标节点的生成')
            #######################################################################################################################
            # 获取目标特征
            Result_Data_dict = Get_Aggregated_Data_From_Online(Spark_Session, tmp_aim_UID_info_dict, Feature_Data_From_Online_Config_dict,
                                                               tmp_all_output_data_batch_dir, 
                                                               Model_Config_dict['Data_Regenerate'], Model_Config_dict['Meta_path_drop_list'],
                                                               Model_Config_dict['Meta_path_Column_drop_dict'],
                                                               Model_Config_dict['Feature_Preprocess_Type_List'])

            All_Meta_Path_Name_list = list(Result_Data_dict['Meta_Path_Feature'].keys())
            Node_Type_to_Feature_len_dict = Result_Data_dict['Node_Type_to_Feature_len'].copy()
            Meta_Path_Column_Type_dict = Result_Data_dict['Meta_Path_Column_Type'].copy()
            
            #######################################################################################################################
            # 根据机器学习格式要求合并特征列
            Result_Data_dict = Concat_Feature_For_ML_Model(Result_Data_dict)
            
            # 进行预测
            tmp_predict_proba = Mlgb.predict(Result_Data_dict['Feature'])
            tmp_predict_proba_pd = pd.DataFrame(tmp_predict_proba, columns = ['Proba'])
            
            # 和target_node文件拼接
            target_node_with_proba_pd = pd.concat([tmp_sub_aim_entity_pd[['Mobile_Node_UID']], tmp_predict_proba_pd], axis=1)
            
            # 存储结果
            target_node_with_proba_pd.to_pickle(tmp_all_output_data_batch_dir + 'Target_Node_with_ML_Proba.pkl')
            
            target_node_with_proba_pd_list.append(target_node_with_proba_pd)
            
            #######################################################################################################################
            # 更新进度
            sample_batch_start = tmp_range_end
            
        break
        
    except Py4JJavaError:
        print('*******************************************************************************')
        
        Spark_Session.stop()
        Spark_Session._instantiatedContext = None
        

        interrupt_time = datetime.now()
        print('pyspark异常中断，时间:', interrupt_time)
        
        # 如果不是0-9点间的中断且重启次数小于20，则进行重启
        if interrupt_time.hour < 9:
            print('0-9点无法运算，故开始等待')

            time_sleep = (8 - interrupt_time.hour)*3600 + (60 - interrupt_time.minute) * 60
            
            print('休眠时间', time_sleep)
            
            time.sleep(time_sleep)
            
            print('已到早上9点，重启运算')
            
        elif spark_restart_count > 20:
            print('重启超过20次，故终止')

            break
            
        else:
            spark_restart_count = spark_restart_count + 1
            print('重启Spark并重新开始运算')
            
        print('*******************************************************************************')
        
end_time = datetime.now()
print(end_time)
print('花费时间:', end_time - start_time)
```

```python
all_target_node_with_proba_pd = pd.concat(target_node_with_proba_pd_list)

print(all_target_node_with_proba_pd.shape)
    
all_target_node_with_proba_pd = all_target_node_with_proba_pd.sort_values(by=['Proba'], ascending = False)

print(all_target_node_with_proba_pd.shape)

localtime = time.strftime("%m-%d-%H:%M", time.localtime())
all_target_node_with_proba_pd.to_pickle(Feature_Data_From_Online_Store_dir + '/' + localtime + '_ML_签约概率.pkl')
```
