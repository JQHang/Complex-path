```python
# 创建spark_session
os.environ['PYSPARK_PYTHON']="/usr/local/anaconda3/bin/python"

def Start_Spark():
    Spark_Session = SparkSession.builder \
                    .appName("kg_spark_for_model") \
                    .enableHiveSupport() \
                    .config("spark.sql.shuffle.partitions", "500") \
                    .config("spark.sql.broadcastTimeout","3600")\
                    .config("spark.driver.memory", "200g") \
                    .config("spark.executor.memory", "40g")\
                    .config("spark.yarn.appMasterEnv.yarn.nodemanager.container-executor.class", "DockerLinuxContainer")\
                    .config("spark.executorEnv.yarn.nodemanager.container-executor.class", "DockerLinuxContainer")\
                    .config("spark.yarn.appMasterEnv.yarn.nodemanager.docker-container-executor.image-name", 
                            "bdp-docker.jd.com:5000/wise_mart_bag:latest")\
                    .config("spark.executorEnv.yarn.nodemanager.docker-container-executor.image-name", 
                            "bdp-docker.jd.com:5000/wise_mart_bag:latest")\
                    .config("spark.sql.crossJoin.enabled", "true")\
                    .config("spark.driver.maxResultSize", "40g")\
                    .config("spark.driver.memory","20g")\
                    .config("spark.sql.autoBroadcastJoinThreshold","-1")\
                    .getOrCreate()
    
    return Spark_Session
```

```python
# 固定随机值
def setup_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    
setup_seed(42)
```

```python
from datetime import datetime
from kg_lib.utils import read_json_config_file
from kg_lib.utils import mkdir
from kg_lib.utils import Log_save

# 数据来源描述
# Output_Columns_Type = "Head_And_Tail"
# Output_Columns_Type = "ALL_Nodes"
Output_Columns_Type = "ALL_Nodes_And_Edges"

# 输出的特征时间数目
Feature_Month_Range = 3

# 采样起始位置
sample_batch_start = 0

# 一批采样的样本总数
sample_batch_size = 2000000

data_source_description_str = ('02_19-23年1月目标战区法人-过去' + str(Feature_Month_Range) + '个月的特征-' + 
                                '每批样本数' + str(sample_batch_size) + '-' + Output_Columns_Type + '格式')

# 目标时间
Aim_Feature_Time = '2023-01-01'

# 日志信息保存文件名
mkdir('../../Log/')
log_file_name = '../../Log/' + data_source_description_str + '-' + datetime.now().strftime("%Y-%m-%d-%H:%M") + '.txt'

Feature_Dataset_Config_file = './kg_config/Feature_Dataset_Config/Feature_Dataset_Config_2023_02_11.json'
Metapath_Config_file = './kg_config/Metapath_Config/Metapath_Config_2023_02_19_for_LHG_or_ML.json'

Feature_Dataset_Config_dict = read_json_config_file(Feature_Dataset_Config_file)
Metapath_Config_dict = read_json_config_file(Metapath_Config_file)
```

```python
from kg_lib.Metapath_Dataloader import get_meta_path
from kg_lib.Node_Feature_Dataloader import get_node_related_features
from kg_lib.utils import mkdir

def get_meta_path_related_pandas_data_and_store_by_source(Spark_Session, sample_source_str, sample_source_dt, tmp_output_columns_type):
    
    # 数据存储位置
    tmp_all_output_data_base_dir = '../../Data/'
    mkdir(tmp_all_output_data_base_dir)
    
    # 设置召回源名称
    tmp_all_output_data_source_dir = (tmp_all_output_data_base_dir + sample_source_str + '/')
    mkdir(tmp_all_output_data_source_dir)
    
    time_range_str = ('Time_Range:' + sample_source_dt)
    tmp_all_output_data_time_range_dir = (tmp_all_output_data_source_dir + time_range_str + '/')
    mkdir(tmp_all_output_data_time_range_dir)
    
    #######################################################################################################################
    # 读取召回源
    tmp_aim_entity_pd = pd.read_pickle('../../Target_Node/全国法人新召回源_去重后_2月.pkl')
    tmp_aim_entity_pd = tmp_aim_entity_pd[['irr_mobile_sha']].copy()
    tmp_aim_entity_pd = tmp_aim_entity_pd.rename(columns = {"irr_mobile_sha": "Mobile_Node_UID"})
    
    # 获得召回源样本总数
    tmp_legal_person_mobile_num = tmp_aim_entity_pd.shape[0]
    print('召回源样本总数:', tmp_legal_person_mobile_num)
    
    for tmp_range_start in range(sample_batch_start, tmp_legal_person_mobile_num, sample_batch_size):
        tmp_index = tmp_range_start//sample_batch_size
        
        tmp_start_time = datetime.now()
        
        tmp_range_end = tmp_range_start + sample_batch_size
        if tmp_range_end > tmp_legal_person_mobile_num:
            tmp_range_end = tmp_legal_person_mobile_num
        
        print('当前处理样本区间', tmp_range_start, tmp_range_end, tmp_index)
        
        tmp_all_output_data_batch_dir = (tmp_all_output_data_time_range_dir + str(tmp_range_start) + '-' + str(tmp_range_end) + '/')
        mkdir(tmp_all_output_data_batch_dir)
        
        tmp_sub_aim_entity_pd = tmp_aim_entity_pd.iloc[tmp_range_start:tmp_range_end, :].reset_index(drop = True)
        
        # 转rdd
        tmp_node_table_schema = StructType([StructField('Mobile_Node_UID', StringType(), True)])
        tmp_aim_entity_rdd = Spark_Session.createDataFrame(tmp_sub_aim_entity_pd, tmp_node_table_schema)
        
        tmp_aim_entity_info_dict = {'Node_Type': 'Mobile_Node',
                           'Data': tmp_aim_entity_rdd,
                           'Monthly_dt': sample_source_dt}

        tmp_output_data_label_file = tmp_all_output_data_batch_dir + 'Target_Node.pkl'
        
        if not os.path.exists(tmp_output_data_label_file):
            tmp_sub_aim_entity_pd.to_pickle(tmp_output_data_label_file)
        
        print('完成目标节点的生成')
        #######################################################################################################################
        # 设定元路径关系表存储文件夹
        tmp_all_output_data_time_range_meta_path_dir = tmp_all_output_data_batch_dir + 'Meta_Path/'
        mkdir(tmp_all_output_data_time_range_meta_path_dir)

        # 设定节点表存储文件夹
        tmp_all_output_data_time_range_node_dir = tmp_all_output_data_batch_dir + 'Node/'
        mkdir(tmp_all_output_data_time_range_node_dir)

        # 从线上表中取元路径数据及对应的节点特征，并保存
        get_meta_path(Spark_Session, tmp_aim_entity_info_dict, Metapath_Config_dict, tmp_all_output_data_time_range_meta_path_dir,
                      tmp_all_output_data_time_range_node_dir, tmp_output_columns_type)

        #######################################################################################################################
        # 设定特征表存储文件夹
        tmp_all_output_data_time_range_feature_dir = tmp_all_output_data_batch_dir + 'Feature/'
        mkdir(tmp_all_output_data_time_range_feature_dir)

        # 从线上表中取特征数据，并保存
        get_node_related_features(Spark_Session, tmp_aim_entity_info_dict, Feature_Dataset_Config_dict, 
                                  tmp_all_output_data_time_range_node_dir, tmp_all_output_data_time_range_feature_dir,
                                  Feature_Month_Range)
        
        #######################################################################################################################
        tmp_end_time = datetime.now()

        print('完成区间', tmp_range_start, tmp_range_end, '内的数据的处理，花费时间：', tmp_end_time - tmp_start_time)
        print('**************************************************************************')
        
    return
```

```python
from datetime import datetime
from py4j.protocol import Py4JJavaError

# 生成输出文件，并记录信息
log_file_f = open(log_file_name, 'w')
sys.stdout = Log_save(sys.stdout, log_file_f)

print('日志输出文件名:', log_file_name)
print('数据输出文件夹:', data_source_description_str)
print('使用的特征配置:', Feature_Dataset_Config_file)
print('使用的元路径配置:', Metapath_Config_file)

start_time = datetime.now()
print(start_time)

spark_restart_count = 0

while True:
    try:
        # 启动spark
        Spark_Session = Start_Spark()
        
        get_meta_path_related_pandas_data_and_store_by_source(Spark_Session, data_source_description_str, Aim_Feature_Time, 
                                                              Output_Columns_Type)
        
        break
        
    except Py4JJavaError:
        print('-------------------------------------------------------')
        
        Spark_Session.stop()
        Spark_Session._instantiatedContext = None
        
        interrupt_time = datetime.now()
        print('pyspark异常中断，时间:', interrupt_time)
        
        # 如果不是0-8点间的中断且重启次数小于10，则进行重启
        if interrupt_time.hour < 8:
            print('0-8点无法运算，故开始等待')

            time_sleep = (7 - interrupt_time.hour)*3600 + (60 - interrupt_time.minute) * 60
            
            print('休眠时间', time_sleep)
            
            time.sleep(time_sleep)
            
            break
        elif spark_restart_count > 10:
            print('重启超过10次，故终止')
            break
            
        else:
            spark_restart_count = spark_restart_count + 1
            print('重启Spark并重新开始运算')
            
        print('-------------------------------------------------------')
        
end_time = datetime.now()
print(end_time)
print(end_time - start_time)
```
