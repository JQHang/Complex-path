# 预处理ML格式的数据

```python
from kg_lib.Get_ComplexPath_Required_Data import Get_ComplexPath_Data_From_Online, Concat_ComplexPath_Feature_For_ML_Model
from kg_lib.Target_Node_Dataloader import get_aim_UID_with_label_rdd, Extend_feature_time_for_aim_UID
from py4j.protocol import Py4JJavaError

spark_restart_count = 0

start_time = datetime.now()
print(start_time)

while True:
    try:
        # 启动spark
        Spark_Session = Start_Spark()

        time_range_str = ('Time_Range:' + str(all_aim_time_range))

        # 根据目标时间创建对应文件夹
        Feature_Data_From_Online_Time_Store_dir = Feature_Data_From_Online_Store_dir + time_range_str + '/'
        mkdir(Feature_Data_From_Online_Time_Store_dir)
        print('目标时间数据存储文件夹', Feature_Data_From_Online_Time_Store_dir)

        # 获取整个目标时间对应的目标点
        tmp_aim_UID_info_dict = get_aim_UID_with_label_rdd(Spark_Session, Label_Data_Config_dict, all_aim_time_range[0], 
                                                           all_aim_time_range[1], Feature_Data_From_Online_Time_Store_dir, True)

        # 基于Feature_Month_Range扩展目标点对应特征时间
        tmp_aim_UID_info_dict = Extend_feature_time_for_aim_UID(Spark_Session, tmp_aim_UID_info_dict, Feature_Month_Range)

        # 获取目标特征
        Result_Data_dict = Get_ComplexPath_Data_From_Online(Spark_Session, tmp_aim_UID_info_dict, Feature_Data_From_Online_Config_dict,
                                                            Feature_Data_From_Online_Time_Store_dir, 
                                                            Model_Config_dict['Data_Regenerate'], Model_Config_dict['Meta_path_drop_list'],
                                                            Model_Config_dict['Meta_path_Column_drop_dict'],
                                                            Model_Config_dict['Feature_Preprocess_Type_List'])
        
        All_Meta_Path_Name_list = list(Result_Data_dict['Meta_Path_Feature'].keys())
        Node_Type_to_Feature_len_dict = Result_Data_dict['Node_Type_to_Feature_len'].copy()
        Meta_Path_Column_Type_dict = Result_Data_dict['Meta_Path_Column_Type'].copy()
        
        break
        
    except Py4JJavaError:
        print('*******************************************************************************')
        
        Spark_Session.stop()
        Spark_Session._instantiatedContext = None
        

        interrupt_time = datetime.now()
        print('pyspark异常中断，时间:', interrupt_time)
        
        # 如果不是0-9点间的中断且重启次数小于20，则进行重启
        if interrupt_time.hour < 9:
            print('0-9点无法运算，故开始等待')

            time_sleep = (8 - interrupt_time.hour)*3600 + (60 - interrupt_time.minute) * 60

            print('休眠时间', time_sleep)

            time.sleep(time_sleep)

            print('已到早上9点，重启运算')

        elif spark_restart_count > 20:
            print('重启超过20次，故终止')

            break
            
        else:
            spark_restart_count = spark_restart_count + 1
            print('重启Spark并重新开始运算')

        print('*******************************************************************************')
        
end_time = datetime.now()
print(end_time)
print('花费时间:', end_time - start_time)
```

# 取出各时间段对应的数据

```python
def Get_Required_ML_Data_by_Time(tmp_aim_UID_info_dict, Result_Data_dict, aim_time_monthly_range_list, delete_useless_rows = True):
    Required_Time_Data_dict = {}
    
    # 获取目标时间内的点对应的index
    tmp_aim_node_pd = tmp_aim_UID_info_dict['Data_pd']
    tmp_required_time_node_index = ((tmp_aim_node_pd['Source_Time'] >= aim_time_monthly_range_list[0].strftime("%Y-%m-%d")) &
                                    (tmp_aim_node_pd['Source_Time'] < aim_time_monthly_range_list[1].strftime("%Y-%m-%d")))
    
    # 如果一个点存在多个月的特征，则得对特征按source_time进行分隔，待优化*
    
    # 获取对应点的标签
    Required_Time_Data_dict['Label'] = Result_Data_dict['Label'][tmp_required_time_node_index].values
    
    # 打印正负样本数目
    print('正样本数:', list(Required_Time_Data_dict['Label']).count(1))
    print('负样本数:', list(Required_Time_Data_dict['Label']).count(0))
    
    # 获取对应点的特征
    Required_Time_Data_dict['Feature'] = Result_Data_dict['Feature'][tmp_required_time_node_index].values
    
    # 保留特征列名
    Required_Time_Data_dict['Columns'] = list(Result_Data_dict['Feature'].columns)
    
    # 删去已经被取出的行的数据
    if delete_useless_rows:
        Result_Data_dict['Label'] = Result_Data_dict['Label'].drop(tmp_required_time_node_index)
        Result_Data_dict['Feature'] = Result_Data_dict['Feature'].drop(tmp_required_time_node_index)
        
    return Required_Time_Data_dict
```

```python
# 根据机器学习格式要求合并特征列
Result_Data_dict = Concat_ComplexPath_Feature_For_ML_Model(Result_Data_dict)

Train_Data_Dict = Get_Required_ML_Data_by_Time(tmp_aim_UID_info_dict, Result_Data_dict, KG_train_time_range_list)
print("训练集维度：", Train_Data_Dict['Feature'].shape) 

Validation_Data_Dict = Get_Required_ML_Data_by_Time(tmp_aim_UID_info_dict, Result_Data_dict, KG_validation_time_range_list)
print("验证集维度：", Validation_Data_Dict['Feature'].shape) 

Test_Data_Dict = Get_Required_ML_Data_by_Time(tmp_aim_UID_info_dict, Result_Data_dict, KG_test_time_range_list)
print("测试集维度：", Test_Data_Dict['Feature'].shape)
```

```python
def top_k_accuracy_score(y_true, y_score, k):
    sorted_pred = np.argsort(y_score)
    sorted_pred = sorted_pred[::-1]
    sorted_pred = sorted_pred[:k]

    hits = y_true[sorted_pred]
    
    return np.sum(hits)/k
```

```python
def evaluate(source_Data_Dict, predict_proba, print_figure = True):
    fpr, tpr, thresholds = roc_curve(source_Data_Dict['Label'], predict_proba)
    roc_auc = auc(fpr, tpr)

    precision, recall, _ = precision_recall_curve(source_Data_Dict['Label'], predict_proba)
    pr_auc = auc(recall, precision)

    top_k_acc_dict = {}
    for tmp_aim_k in [100, 500, 1000, 2000, 5000, 10000]:
        top_k_acc_dict[tmp_aim_k] = top_k_accuracy_score(source_Data_Dict['Label'], predict_proba, k = tmp_aim_k)
    
    if print_figure:
        plt.plot(fpr, tpr, 'b', label = 'Val AUC = %0.3f' % roc_auc)
        plt.plot([0, 1], [0, 1],'r--')
        plt.xlim([0, 1])
        plt.ylim([0, 1])
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plt.show()
        
        plt.plot(precision, recall)
        plt.xlim([0, 1])
        plt.ylim([0, 1])
        plt.ylabel('recall')
        plt.xlabel('precision')
        plt.show()
    
    print('roc_auc',roc_auc)
    print('pr_auc',pr_auc)
    print(top_k_acc_dict)

    return
```

```python
evaluate(Train_Data_Dict, train_predict_proba)
evaluate(Validation_Data_Dict, validation_predict_proba)
evaluate(Test_Data_Dict, test_predict_proba)
```
